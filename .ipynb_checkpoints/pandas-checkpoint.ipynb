```python
import pandas as pd
```

```python
movies_df = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies_df.head()
```

```python
# 3 null values in content_rating column

movies_df.info()
```

How to sort a series

```python
# how to sort a series with sort_values(). Returns in ascending order by default

# sort_values doees not change the underlying data, can do it inplace

movies_df.title.sort_values()

# or 

movies_df['title'].sort_values()
```

```python
type(movies_df.title.sort_values())
# returns a pandas series
```

```python
# sort in descending order

movies_df.title.sort_values(ascending=False)
```

how to sort a dataframe by one series/column

```python
# sorting a dataframe by a series

# pass it the name of the column you want to sort by
# can also specify ascending
# dataframe method that returns dataframe
# does not affect undelrying data

movies_df.sort_values(by='title') 
```

```python
type(movies_df.sort_values(by='title'))
```

how to sort a dataframe by multiple series/columns

```python
# sort a dataframe by multiple columns, pass a list of strings

movies_df.sort_values(by=['content_rating', 'duration']).head()
```

```python
# sort a dataframe by multiple columns, with different orders (one ascending one descending)
    
movies_df.sort_values(by=['content_rating', 'duration'], ascending=[True,False], inplace=True)
```

sort dataframe by index

```python
# can sort by index, use sort index method

movies_df.sort_index()
```

```python
# how to find the largest and smallest values from a dataframe

# use nlargest method

movies_df['duration'].nlargest(10)
```

better way to find the largest values from a dataframe

```python
# get entire dataframe with the 10 largest duration numbers

movies_df.nlargest(10, 'duration')
```

better way to find the smallest values from a dataframe

```python
# get entire dataframe with the 10 largest duration numbers

movies_df.nsmallest(10, 'duration')
```

```python
movies_df.isna().sum()
```

```python
# other notes:
# na_position parameter for NaN values order
# ignore_index parameter to ignore the index if set True, otherwise will maintain same index
# can add function to key parameter, should be vectorized, expects series and returns series with same shape as input, which is each column passed in to by parameter
# can sort columns by row values using axis=1
# can sort columns by column index using sort_index() and axis=1
```

indexing a dataframe: using callable function with loc method

```python
# callable function for loc method

movies_df.loc[lambda df: df.content_rating == 'APPROVED', :]
```
